```python


c=["#5cb85c","#5bc0de","#d9534f"]

fig = plt.figure()
ax = fig.add_subplot()
ax = df.plot(kind='bar', width=0.8, figsize = (20,8), color=c, fontsize=14)

ax.set_title("Percentage of Respondents Interests in Data Science Area", fontsize=16)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')

ax.spines['left'].set_color('black')
ax.set_facecolor("white")

ax.set_yticks([])
plt.grid(b=None)

ax.set_xlabel('', fontsize=14)
ax.set_ylabel('', fontsize=14)
plt.legend(prop={'size': 14})


totals = []

for i in ax.patches:
    totals.append(i.get_height())
    
total = sum(totals)

for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+0.0, i.get_height()+40.0, 
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=14,
                color='dimgrey')
    
```

```python
path_sf = 'https://cocl.us/sanfran_crime_dataset/Police_Department_Incidents_-_Previous_Year__2016_'
df_sf = pd.read_csv(path_sf)
```

```python
df_sf.head()
```

```python
df_sf.shape
```

```python
df_sf.drop(df_sf[df_sf.Category == 'NON-CRIMINAL'].index, inplace = True)
```

```python
df_sf.shape
```

```python
df_sfn = df_sf.groupby('PdDistrict', axis = 0).count()
```

```python
df_sfn.head()
```

```python
df_sfn.reset_index(inplace=True)
```

```python

df_sfn.rename(columns = {'PdDistrict':'Neighborhood', 'IncidntNum': 'Count'}, inplace = True)
```

```python
df_sfn = df_sf.groupby('PdDistrict', axis = 0).count()
```

```python
column_lst = list(df_sfn.columns.values)
```

```python
column_lst
```

```python
df_sfn.columns = ['Neighborhood','Count','Category','Descript','DayOfWeek','Date', 'Time','Resolution','Address','X','Y','Location','PdId']
```

```python
df_sfn.index = [9, 0, 8, 4, 1, 2, 6, 3, 7, 5]
```

```python
!wget --quiet  https://cocl.us/sanfran_geojson/san-francisco.json -O san-francisco.json
```

```python
san_geo = r'san-francisco.json'
```

```python
latitude = 37.77
longitude = -122.42
```

```python

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
sanfran_map.choropleth(
    geo_data=san_geo,
    data=df_sfn,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

# display map
sanfran_map

```
